```python
data['MINIMUM_PAYMENTS'] = data['MINIMUM_PAYMENTS'].fillna(data['MINIMUM_PAYMENTS'].mean())
data['CREDIT_LIMIT'] = data['CREDIT_LIMIT'].fillna(data['CREDIT_LIMIT'].mean())
```

```python
data.drop(columns = ['CUST_ID'],axis = 1, inplace = True)
```

```python
data.columns
```

```python
plt.figure(figsize = (30,45))
for i, col in enumerate(data.columns):
    if data[col].dtype != 'object':
        ax = plt.subplot(9, 2, i+1)
        sns.distplot(data[col], ax = ax)
        plt.xlabel(col)

plt.tight_layout()
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(data.corr(),annot=True)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
scaled_data
```

```python
from sklearn.cluster import KMeans
```

```python
inertia = []
range_val = range(1,12)
for i in range_val:
    kmean = KMeans(n_clusters=i)
    kmean.fit_predict(pd.DataFrame(scaled_data))
    inertia.append(kmean.inertia_)
plt.plot(range_val,inertia,'bx-')
plt.xlabel('Value of K ')
plt.ylabel('Inertia')
plt.title('The Elbow Method using Inertia')
plt.show()
```

```python
kmeans_model=KMeans(n_clusters = 4 , random_state = 1)
kmeans_model.fit(scaled_data)
```

```python
cluster_labels = kmeans_model.labels_
```

```python
simpleClusterInfo = pd.Series(kmeans_model.labels_).value_counts().sort_index()
print(simpleClusterInfo) 
```

```python
set(kmeans_model.labels_)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(scaled_data[:, 2], scaled_data[:, 3], c=cluster_labels, cmap='viridis')
plt.title("Scatter Plot of K-means Clusters (Scaled Data)")
plt.xlabel("Feature 1 (Scaled)")
plt.ylabel("Feature 2 (Scaled)")
plt.show()
```

```python
cluster_centers = pd.DataFrame(data = kmeans_model.cluster_centers_,columns = [data.columns])

cluster_centers = scaler.inverse_transform(cluster_centers)
cluster_centers = pd.DataFrame(data=cluster_centers , columns = [data.columns])
cluster_centers
```

```python
cluster_data = pd.concat([data,pd.DataFrame({'Cluster' : kmeans_model.labels_})],axis = 1)
cluster_data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = cluster_data.drop(["Cluster"],axis = 1)
y = cluster_data[["Cluster"]]
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size= 0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(criterion= "entropy")
model.fit(x_train , y_train)
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report , accuracy_score ,f1_score
from sklearn.metrics import confusion_matrix , roc_auc_score ,roc_curve
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(classification_report(y_test ,y_pred))
```

```python
y_pred = model.predict(x_test)

km = confusion_matrix(y_test ,y_pred)

plt.figure(figsize=(8,6))
sns.heatmap(km , annot=True ,fmt='d',cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```
